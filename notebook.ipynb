# The Effect of US Presidents on EUR/USD Movement

## The Euro/Dollar Pair
Whatever asset you decide to trade, one of the first lesson your mentor will teach you is to *look at the Euro and Dollar long term trend*. Traders persistently try to predict the EURO/USD long term movements, sometimes ending up in circles of uncertainty.

![](https://www.tradefw.com/wp-content/uploads/2018/12/EUR-USD-currency-pair-1024x274.jpg)

It wasn't until 1999 that the euro really began its journey, when 11 countries (Austria, Belgium, Finland, France, Germany, Ireland, Italy, Luxembourg, the Netherlands, Portugal and Spain) fixed their exchange rates and created a new currency with monetary policy passed to the European Central Bank. The currency has withstood the test of time, and today the Euro is over 20 years old. US Financial data providers now ask at the start of every year – what will be the direction of the EURO/USD pair?

This project will attempt to use a storytelling perspective to unfold the relationship between US Election results and the the EURO/USD pair.

## Dataset Information

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. It contains date and Euro rates corresponding to 40 currencies all over the world. This data has been put together and uploaded on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) by a user called *Daria Chemkaeva*. The data is constantly updated from its original source — the European Central Bank.


### What exactly is exchange rate?
An exchange rate is the rate at which a country's currency will be exchanged for another country's currency. For example, If the exchange rate of the euro to the US dollar is **1.5, you will get 1.5 US dollars if you pay 1.0 euro** (one euro has more value than one US dollar at this exchange rate).

### Initial Notes
- *There are 5699 rows and 41 columns in the dataset. The first column corresponds to time, while the others correspond to 40 different currencies.*
- *17 columns have null values. In fact, over 50% of values in some columns are null. Notable examples are the Greek drachma and Slovanian tolar columns.*
- *Since we are dealing with currencies, many of these columns should be numeric. However, only three of the currency columns are stored in the correct data type.*
- *The period/Unit column should be stored as a DateTime object rather than a string/object type.*
- *Column names do not follow the Python snake case convention. They also contain extra characters that might make them challenging to work with.*

In the next few steps we will be correcting some of the irregularities that we have outlined. We are only concerned about the EUR/USD pair, hence, we will direct our cleaning towards the required columns - `Period/Unit` and `[US dollar ]`.

## Data Cleaning
Our cleaning will follow three simple steps. First we'll rename the two columns: `Period/Unit` and `[US dollar ]`. Next, we'd extract the columns into a new dataframe called `usd_data`. Finally, we will convert the time column and the dollar column into the right data types:

````python
# Rename relevant columns
exchange.rename(columns={'Period\\Unit:': 'time', '[US dollar ]':'us_dollar'}, inplace=True)

# Extract relevant columns
usd_data = exchange.loc[:,['time', 'us_dollar']]

# Datatype conversion
usd_data['time'] = pd.to_datetime(exchange['time'])

usd_data.sort_values(by='time', inplace=True)
usd_data.reset_index(drop=True, inplace=True)

# Examine the us_dollar column
print( '\033[1m'+'us_dollars: Unique values'+  '\033[0m')
usd_data['us_dollar'].value_counts()
````

It appears that we have some missing values recorded as **'-'** in the `us_dollar` column. These values can cause our conversion process to throw up errors. They can also affect the results of our analysis. As a result, we are going to eliminate them from our data:

````python
usd_data = usd_data.query("us_dollar != '-'").copy()

usd_data['us_dollar'] = usd_data['us_dollar'].astype(float)
print( '\033[1m'+'Cleaned Column Datatypes'+  '\033[0m')
usd_data.dtypes
````

## Exploratory Visualization
Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate:

````python
plt.plot(usd_data['time'], usd_data['us_dollar'])
plt.show()
````

Its quite challenging to make sense of this graph. We see many small wiggles — rather than seeing a smooth line. These are the visual representation of the daily variations in exchange rate. The rate goes up and down from day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to show only the long-term trends, we can use the moving average(rolling mean). To calculate the rolling mean over any period of time, we can use the handy `pandas.Series.rolling().mean()` method:

````python
plt.figure(figsize=(12,8))
plt.subplot(3,2,1)
plt.plot(usd_data['time'], usd_data['us_dollar'])
plt.title('Original', fontweight='bold')

for i, roll in zip(range(2,7), (7, 30, 50, 100, 365)):
    plt.subplot(3,2,i)
    plt.plot(usd_data['time'], usd_data['us_dollar'].rolling(roll).mean())
    plt.title('Rolling window: {}'.format(roll), fontweight='bold')

plt.tight_layout()
plt.show()
````

The 30 day rolling window appears to be a sweet spot. We can see detailed fluctuations in exchange rates, without losing track of long term trends. For most of our analysis, we will stick to the 30 day rolling window. Let's add this column to our dataframe:

````python
usd_data['rolling_usd'] = usd_data['us_dollar'].rolling(30).mean()

print( '\033[1m'+'Columns in usd_data'+  '\033[0m')
num = 1
for column in usd_data.columns:
    print('{}: {}'.format(num, column))
    num+=1
````


## Our Story: The US Elections Influence on EURO/USD Rates
Without getting so deep in politics, the US ruling party may influence the EUR/USD pair. There might be a correlation between a political party's agenda and the USD price. Every ruling party has a different economic plan. In general, Republicans tend to focus inwards, control tax levels, revise agreements with other countries, and increase military spending. Democrats focus on external expansion and decreasing military spending. These policies could stimulate or negatively impact the US economy.

````python
usd_data['year'] = usd_data['time'].dt.year

# Isolating the different presidential periods
clinton = usd_data.query("year < 2001")
bush = usd_data.query("(year >=2001) & (year <2009)")
obama = usd_data.query("(year >=2009) & (year <2017)")
trump = usd_data.query("(year >=2017) & (year <2021)")

# Creating the main plot 
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(11.5,6))
ax.plot(usd_data['time'], usd_data['us_dollar'].rolling(30).mean(),
       color='#fff', linewidth=2)

# Customizing ticks and spines
ax.tick_params(axis='x', colors='grey', labelsize=14)
ax.tick_params(axis='y', colors='grey', labelsize=14)
for loc in ('top', 'left', 'right', 'bottom'):
    ax.spines[loc].set_visible(False)
ax.tick_params(left=False, bottom=False)
ax.set_xticks([])
ax.set_yticks([0.6, 0.9, 1.2, 1.5])
ax.set_yticks([])

# Color coding the ruling parties 
ax.axvspan(xmin=10550, xmax=11310, color='#5768AC', ymin=0.15)
ax.axvspan(xmin=11330, xmax=14200, color='#FA5A50',ymin=0.15)
ax.axvspan(xmin=14220, xmax=17110, color='#5768AC', ymin=0.15)
ax.axvspan(xmin=17130, xmax=18600, color='#FA5A50', ymin=0.15)

# Texts, legends and line markers
xtexts = ['Bush succeeds Clinton', 'Obama succeeds Bush', 'Trump succeeds Obama']
years=(2001, 2009, 2017)
for coord, year, text in zip((11330, 14200, 17130, 18630), years, xtexts):
    ax.axvline(x=coord, color='black', linewidth=1, alpha=0.5, ymin=0.15)
    ax.text(coord-180, 1.76, s=year, size=14)
    ax.text(coord-800, 1.70, s=text, size=14, color='grey')
    
ax.axhline(y=1.18, color='#222', linewidth=1, xmin=0.04, xmax=0.96)
ax.text(x=10000, y=1.18, s='1 Euro = 1.2USD', size=14, backgroundcolor='#333', color='w')

ax.text(10700, 0.43, s='Democratic rule', size=13)
ax.text(10700, 0.36, s='Republican rule', size=13)
ax.axhline(y=0.45, linewidth=3, xmax=0.04, color='#5768AC')
ax.axhline(y=0.38, linewidth=3, xmax=0.04, color='#FA5A50')

# Title and subtitle
ax.text(10000, 2.0, s= 'Did the US Elections Influence the Euro/USD Rate?', 
        size=28, weight='bold')
ax.text(10000, 1.9, s= '''USD closes the gap during democratic periods, but Euro wins during republican periods''',
        size=18)

# Signature
ax.text(10000, 0.2, '©ISRAEL OGUNMOLA' + ' '*74 + 'Source: EUROPEAN CENTRAL BANK',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=15)

plt.show()
````

It is evident from the chart how much the US ruling party influences the EURO/USD pair. The trendlines are falling during the democratic periods and they are rising during the Republican periods. In other words, Republican fiscal policy builds upon the weaker dollar, while the stronger dollar results from the Democrat's fiscal policy.
